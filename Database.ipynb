# Variables independientes

Explicare que representa cada variable, igual lo explico en el PDF adjunto.

Los datos fueron sacados de la página de eurostat, donde se almacenan datos de diversos temas sobre los paises de europa y los dividen por año, sexo, edad, etc.

En este caso, lo dividi por año porque la situación general de un país puede cambiar dependiendo de su situación.

Filas:  Cada fila representa un país y un año, es decir, la situación en que se encontraba un país en concreto un año en concreto. Esto es relevante porque de aquí en adelante me referire a las filas como paises ya que generalmente no hay una diferencia muy grande entre los distintos años, por eso hago la aclaración de que son ambos, país y año.

```python
Deuda = pd.read_csv("Deudas.csv.gz")
```

Deuda: Personas con moras (hipoteca o alquiler, facturas de servicios públicos o compra a plazos), porcentaje.

```python
Desem = pd.read_csv("Desempleo.csv.gz")
```

Desempleo: Desempleo de larga duración de 15 a 74 años, porcentaje %.

```python
Edu = pd.read_csv("NivelEducativo_Uni.csv.gz")
```

Educación Universitaria: Personas con nivel educativo universitario de 15 a 64 años, porcentaje %.

```python
SD = pd.read_csv("Pricioneros-SenteciaDictada.csv.gz")
```

Prisioneros Sentenciados: Personas en prisión con sentencia dictada, por cada 100,000.

```python
PJ = pd.read_csv("Prisioner-PreJuicio.csv.gz")
```

Prisioneros Prejuicio: Personas en prisión preventiva, por cada 100,000 habitantes.

```python
RP = pd.read_csv("RiesgoPobreza.csv.gz")
```

Riesgo de Pobreza: Personas en un estado persistente de riesgo pobreza, por porcentaje.

```python
PS = pd.read_csv("ProteccionSocial.csv.gz")
```

Protección social: Principales gastos en protección social, Euro por habitante.

```python
PR = pd.read_csv("PrivacionRecursos.csv.gz")
```

Privación de Recursos: Tasa de privación material y social de 16 a 64 años, por porcentaje %.

```python
Inm = pd.read_csv("Inmigración.gz")
```

Inmigración: Número de inmigrantes registrados anualmente.

```python
Emi = pd.read_csv("Emigración.csv.gz")
```

Emigración: Número de emigrantes registrados anualmente.

```python
Robos = pd.read_csv("Robos.csv.gz")
```

Robos: Robos registrados por la policía por cada 100,000 habitantes. Este incluye únicamente aquellos robos donde NO hubo intimidación o violencia involucrada, tampoco incluye allanamiento, o fraude.

```python
Robos_violencia = pd.read_csv("Robos-violencia.csv.gz")
```

Robos con violencia: Cualquier tipo de robo en el que hubiera violencia física involucrada registrada por la policía por cada 100,000 habitantes.

```python
Ase = pd.read_csv("Homicidios.csv.gz")
```

Homicidios: Homicidios intencionales registrados por la policía por cada 100,000 habitantes.

```python
IntH = pd.read_csv("Intento_H.csv.gz")
```

Intento de Homicidio: Intentos de homicidios intencionales registrados por la policia por cada 100,000 habitantes.

```python
VS = pd.read_csv("ViolenciaSexual.csv.gz")
```

Violencia Sexual: Casos de violencia sexual registrados por la policía por cada 100,000 habitantes.

```python
Corr = pd.read_csv('Corrupción.csv.gz')
```

Corrupción: Casos de corrupción registrados por la policía por cada 100,000 habitantes.

```python
dfs = [Deuda, Desem, Edu, SD, PJ, RP, PS, PR, Inm, Emi, Robos, Robos_violencia, Ase, IntH, VS, Corr]
```

```python
for df in dfs:
    print(df)
```

```python
columnas_a_conservar = ['geo', 'TIME_PERIOD', 'OBS_VALUE']
nombres_reales = ['Deuda', 'Desempleo', 'Educación Universitaria', 'Prisioneros Setenciados', 'Prisioneros Prejuicio', 'Riesgo de Pobreza', 'Protección Social', 'Privación de Recursos', 'Inmigración', 'Emigración', 'Robos', 'Robos con violencia', 'Homicidios', 'Intento de Homicidio', 'Violencia Sexual', 'Corrupción']
```

```python
for i, df in enumerate(dfs):
    df = df[columnas_a_conservar]
    df.columns = ['geo', 'TIME_PERIOD'] + [nombres_reales[i]]
    dfs[i] = df
```

```python
for df in dfs:
    print(df)
```

```python
merged_df = dfs[0]  
for df in dfs[1:]:
    merged_df = merged_df.merge(df, on=['geo', 'TIME_PERIOD'], how='outer')
```

```python
merged_df
```

```python
df_Completo = merged_df.dropna()
```

```python
df_Completo
```

```python
df_Completo.to_csv("df_Completo.csv", index=False)
```
