```python
df.hist(['Deuda'], bins=15)
plt.title('Deudas')
plt.xlabel('Porcentaje de Deudas')
plt.ylabel('Países divididos por año')
```

Esta gráfica muestra el porcentaje de personas sumidas en deudas que tiene un país, estas pueden ser hipotecas, facturas u otras.

Vemos que la mayoria de los países tienen alrededor del 10% de sus habitantes sumidos en deudas.

Tambien existen casos en el que el país se acerca peligrosamente al 50%, pero ningúno realmente lo pasa.

```python
df.hist(['Desempleo'], bins=15, color='darkred')
plt.title('Desempleo')
plt.xlabel('Porcentaje de Desempleo')
plt.ylabel('Países divididos por año')
```

Esta gráfica muestra el porcentaje de desempleo que tienen los distintos paises.

Observamos como la gran mayoria de los datos estan entre 20-45%, lo cual, personalmente lo encuentro preocupante, especialmente considerando que apartir del 65% es cuando ya hay una reducción verdaderamente significativa y el que existen países con una tasa de desempleo del 80%.

```python
df.hist(['Educación Universitaria'], bins=15, color='darkgreen')
plt.title('Educación Universitaria')
plt.xlabel('Porcentaje de Educación Universitaria')
plt.ylabel('Países divididos por año')
```

Esta gráfica nos muestra el porcentaje de personas con un nivel educativo de nivel universitario que tiene cada país.

Lo primero que me sorprende de esta gráfica es que ningún país logre llegar al 50% de habitantes graduados de la universidad.

La gráfica esta bastante distribuida entre el 20 y 40%, destacando entre el 30 y 35%, teniendo un pico importante justo despues de esta.

```python
df.hist(['Prisioneros Setenciados'], bins=15, color='silver')
plt.title('Prisioneros Setenciados')
plt.xlabel('Número de Prisioneros Setenciados, por cada 100,000 habitantes')
plt.ylabel('Países divididos por año')
```

Estos son los prisiones cuya sentecia ya fue dictada.

Vemos que la mayoria de los países no tienen más de 100 personas setenciadas por cada 100,000, que seria alrededor de 10%, lo cual, podria ser señal de un bajo indice de criminalidad o de un sistema de justicia ineficiente.

Tambien vemos que hay un par de picos pequeños más, uno entre 100 y 150, y otro entre 150 y 200.

```python
df.hist(['Prisioneros Prejuicio'], bins=15, color='orange')
plt.title('Prisioneros Pre-juicio')
plt.xlabel('Número de Prisioneros Pre-juicio, por cada 100,000 habitantes')
plt.ylabel('Países divididos por año')
```

Aquí estan las personas que estan en prisión de manera preventiva, es decir, que estan acusadas de haber cometido un crimen, pero un juicio aún no se ha emitido y para evitar que escapen, estos son enviados a prisión

Observamos como la mayoria de países solo tienen 20 perosnas en prisión preventiva por cada 100,000 con bastantes otros teniendo numeros similares.

La mayor cantidad que ha llegado a tener un país de prisioneros en espera de jucio son poco más de 70 por cada 100,000, pero siguen siendo muy pocos.

```python
df.hist(['Riesgo de Pobreza'], bins=15, color='black')
plt.title('Riesgo de Pobreza')
plt.xlabel('Porcentaje de Riesgo de Pobreza')
plt.ylabel('Países divididos por año')
```

En esta gráfica muestra el pocertaje de personas que estan en un persistente riesgo de pobreza.

La gráfica nos enseña que la mayoria de los paises tienen un riesgo de pobreza entre 5 a 16%, pero tienen 2 picos principales. Uno y el más grande entre 7 y 8% y el segundo en 15%.

Desde mi perspectiva un 10% me parece un número apenas tolerable. Y cualquier número superior me parece alarmante porque, recordemos que estos son riesgos persistentes, gente que esta luchando constantemente con la pobreza y a quienes un evento que pueda dañar la economia como lo fue el covid-19 los llevaran a la pobreza.

```python
df.hist(['Protección Social'], bins=15, color='purple')
plt.title('Protección Social')
plt.xlabel('Protección Social, Euro por cada habitante')
plt.ylabel('Países divididos por año')
```

Los sistemas de protección social ayudan a las personas pobres y vulnerables a hacer frente a las crisis, encontrar empleo e invertir en la salud y la educación.

Lo que vemos en la gráfica es cuanto invierten los países en protección social, o cuanto gastan para rescatar a las personas de la pobreza, dividida por cuanto gastan (en Euros) por persona.

Por lo que se ve, los países no suelen gastar más de 5000€ por persona y teniendo un segundo pico un poco más pequeño entre 10k-15k.

```python
df.hist(['Robos'], bins=15, color='darkseagreen')
plt.title('Robos Registrados por la policia')
plt.xlabel('Número de robos por cada 100,000 habitates')
plt.ylabel('Países divididos por año')
```

La mayoria de los paises no tienen más de 2,000 registrados anualmente por cada 100,000 habitantes siendo el maximo de 4,000.

La gráfica tiene 3 picos, el primero y más grande esta en los 500 y tiene bastantes barras a su alrededor, el segundo se esta poco despues de los 1,500 y tiene una barra un poco más pequeña en los 1,500, y la tercera y más chica esta entre los 2,000 y los 2,500.

Tiene sentido que este sea el más el que más registros tiene, ya que es aquel al que la gente se siente más tentado a cometer al sentir que es el crimen que menos daña a la gente y ser el más facil de cometer.

```python
df.hist(['Robos con violencia'], bins=15, color='darkolivegreen')
plt.title('Robos con violencia registrados por la policia')
plt.xlabel('Número de robos con violencia por cada 100,000 habitates')
plt.ylabel('Países divididos por año')
```
