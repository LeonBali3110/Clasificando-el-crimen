# GridSearch

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import silhouette_score
```

```python
parametros_Agglomerative = {
    'n_clusters': [3],
    'linkage': ['ward', 'complete', 'average', 'single']
}

parametros_Kmeans = {
    'n_clusters': [7],
    'init': ['k-means++', 'random'],
    'algorithm': ['auto', 'full', 'elkan']
}

parametros_Birch = {
    'n_clusters': [2, 3, 7],
    'threshold': [0.3, 0.5, 0.7],
    'branching_factor': [25, 50]
}

parametros_Affinity = {
    'damping': [0.5, 0.7, 0.9],
}

parametros_Shift = {
    'bandwidth': [bandwidth],
    'bin_seeding': [True, False]
}

parametros_Spectral = {
    'n_clusters': [2, 3, 7],
    'eigen_solver': ['arpack', 'lobpcg', 'amg'],
    'assign_labels': ['kmeans', 'discretize', 'cluster_qr']
}

parametros_Mini = {
    'n_clusters': [2, 3, 7],
    'init': ['k-means++', 'random']
}
```

```python
Modelos = [
    ('Agglomerative', AgglomerativeClustering(), parametros_Agglomerative),
    ('Kmeans', KMeans(random_state=0), parametros_Kmeans),
    ('Birch', Birch(), parametros_Birch),
    ('Affinity', AffinityPropagation(random_state=0), parametros_Affinity),
    ('Shift', MeanShift(), parametros_Shift),
    ('Spectral', SpectralClustering(random_state=0), parametros_Spectral),
    ('MiniBatch', MiniBatchKMeans(random_state=0), parametros_Mini)
]
```

```python
resultados_modelos = {}

for nombre, modelo, parametros in Modelos:
    print(f"Entrenando modelo: {nombre}")
    
    def silhouette_scorer(estimator, X):
        labels = estimator.fit_predict(X)
        return silhouette_score(X, labels)
    
    GS = GridSearchCV(estimator=modelo,
                      param_grid=parametros,
                      scoring=silhouette_scorer,
                      cv=4,  
                      verbose=4)
    GS.fit(df_1)
    
    resultados_modelos[nombre] = {
        "mejores_parametros": GS.best_params_,
        "mejor_resultado_silhouette": GS.best_score_,
        "mejor_modelo": GS.best_estimator_
    }
    print("\n" + "="*40 + "\n")
```

```python
for nombre_modelo, resultados in resultados_modelos.items():
    print(f"Resultados para {nombre_modelo}:")
    print(f"Mejores parámetros: {resultados['mejores_parametros']}")
    print(f"Mejor resultado para silhouette: {resultados['mejor_resultado_silhouette']}")
    print(f"Mejor modelo: {resultados['mejor_modelo']}")
    print("\n" + "="*40 + "\n")
```

```python
for nombre_modelo, resultados in resultados_modelos.items():
    plt.barh(nombre_modelo, resultados['mejor_resultado_silhouette'], color='lightgreen')
    plt.xlabel('silhouette Score')
    plt.title('Comparación de silhouette Score entre Modelos')
```

```python
nombres_modelos = list(resultados_modelos.keys())

fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.4

posiciones = np.arange(len(nombres_modelos))
posiciones_resultados = posiciones + bar_width

ax.barh(posiciones, resultados['mejor_resultado_silhouette'], height=bar_width, color='lightgreen', label='Parametros GridSearch')

ax.barh(posiciones_resultados, [clustering_silhouette, kmeans_silhouette, Birch_silhouette, Affinity_silhouette, Shift_silhouette, Spectral_silhouette, MiniBatch_silhouette], height=bar_width, color='teal', label='Parametros Predeterminados')

ax.set_yticks(posiciones + bar_width / 2)
ax.set_yticklabels(nombres_modelos)
ax.set_xlabel('Silhouette Score')
ax.set_title('Comparación de Modelos')
ax.legend()
```

Parece que los nuevos parametros han hecho mejorar el puntaje significativamente, pero vamos a verificar esto.
